```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=['bmw','audi','mersedies','tata']
y=[7,8,9,6]
```

```python
plt.xlabel("cars",fontsize=15)
plt.ylabel("values",fontsize=10)
plt.title("cars vs value")

plt.bar(x,y) 

plt.show()
```

```python
plt.barh(x,y)
```

```python
plt.bar(x,y,width=0.5,color='r')
```

```python

col=['r','g','b','k']

plt.bar(x,y,color=col)
```

```python
plt.bar(x,y,width=0.4,edgecolor='r',linewidth=2,linestyle=":")
```

```python
plt.bar(x,y,alpha=0.2)
```

```python
x=['bmw','audi','mersedies','tata']
y=[7,8,9,6]
z=[4,3,6,0.2]


plt.bar(x,y,label="popularity")
plt.bar(x,z,label="hates")
plt.legend()

plt.show()
```

```python
p = np.arange(len(x))

width=0.2

p1=[j+width for j in p]

plt.bar(p,y,label="popularity",width=width)
plt.bar(p1,z,label="hates",width=width)

plt.xticks(p+width/2,x,rotation=20)

plt.legend()
plt.show()
```
